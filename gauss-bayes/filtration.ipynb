```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel (r'data/walking-priya.xls')
print (df)
```

```python
absavg= df[['Time (s)','Absolute acceleration (m/s^2)']]
absavg.plot(x='Time (s)', y='Absolute acceleration (m/s^2)')
plt.show()
```
